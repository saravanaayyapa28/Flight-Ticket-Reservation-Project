```python
class Booking:
    def __init__(self, booking_id, operation, flight, seat_class, total_seats, meal):
        self.id = booking_id
        self.operation = operation
        self.flight = flight
        self.seat_class = seat_class
        self.total_seats = total_seats
        self.meal = meal

    def BookingID(self):
        print("Booking Id:", self.id)
        print("Flight Number:", self.flight)
        r = []  # List to store seat numbers
        tsum = 0
        seats101=0
        seats102=0
        
        if self.seat_class == 'EC':
            if self.flight == 101:
                available_seats = 19 - (18 - (self.total_seats + seats101))
                for x in range(available_seats, available_seats + self.total_seats):
                    r.append(x)
            elif self.flight == 102:
                available_seats = 19 - (18 - (self.total_seats + seats102))
                for x in range(available_seats, available_seats + self.total_seats):
                    r.append(x)
            print("Economy Class:", ','.join(map(str, r)))

        elif self.seat_class == 'BC':
            if self.flight == 101:
                available_seats = 7 - (6 - (self.total_seats + seats101))
                for x in range(available_seats, available_seats + self.total_seats):
                    r.append(x)
            elif self.flight == 102:
                available_seats = 7 - (6 - (self.total_seats + seats102))
                for x in range(available_seats, available_seats + self.total_seats):
                    r.append(x)
            print("Business Class:", ','.join(map(str, r)))

        if self.meal == "Y":
            print("Meals Required: Yes")
            meal_cost = 1200 if self.seat_class == 'EC' else 2200
            tsum = meal_cost * self.total_seats
        elif self.meal == "N":
            print("Meals Required: No")
            meal_cost = 1000 if self.seat_class == 'EC' else 2000
            tsum = meal_cost * self.total_seats

        print("Total Cost:", tsum)
        
        if self.meal == "Y":
            print("Meals Required: Yes")
        elif self.meal == 'N':
            print("Meals Required: No")

        if self.seat_class == 'EC':
            if self.meal == "Y":
                for y in r:
                    tsum += 1200
            elif self.meal == "N":
                for y in r:
                    tsum += 1000
        elif self.seat_class == 'BC':
            if self.meal == "Y":
                for y in r:
                    tsum += 2200
            if self.meal == "N":
                for y in r:
                    tsum += 2000

        print("Total Cost:", tsum)


class Cancellation(Booking):
    def cancellation(self):
        cancel_fees = 0
        print("Booking Id:", self.id)
        print("Flight Number:", self.flight)
        print("Cancelled")
        for _ in range(self.total_seats):
            cancel_fees += 200
        print("Total Cost:", cancel_fees)


while True:
    a = int(input("Enter Booking ID: "))
    b = input("Enter Operation (B for booking, C for cancellation): ")
    c = int(input("Enter Flight Number: "))
    d = input("Enter Seat Class (EC for Economy Class, BC for Business Class): ")
    e = int(input("Enter Total Seats: "))
    f = input("Enter Meal Preference (Y for Yes, N for No): ")

    if b == "B":
        test = Booking(a, b, c, d, e, f)
        test.BookingID()
    elif b == 'C':
        can = Cancellation(a, b, c, d, e, f)
        can.cancellation()
    else:
        print("Invalid operation. Please enter 'B' for booking or 'C' for cancellation.")

    proceed = input("Do you want to continue (Y/N)? ")
    if proceed.upper() != 'Y':
        break
```
