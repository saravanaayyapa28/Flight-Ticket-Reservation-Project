```python
class Flight:
    
    print('='*25)
    print('welcome to air india')
    print('='*25)
class Customer:
    def __init__(self, customer_id, name):
        self.customer_id = customer_id
        self.name = name

class Flight:
    def __init__(self, flight_number):
        self.flight_number = flight_number
        self.business_seats = 6
        self.economy_seats = 12
        self.base_price_economy = 1000
        self.base_price_business = 2000
        self.current_price_economy = self.base_price_economy
        self.current_price_business = self.base_price_business
        self.bookings = {}

    def book_ticket(self, booking_id, seat_type, num_tickets, meal=False, customer_details=None):
        if seat_type == 'business':
            if self.business_seats >= num_tickets:
                price = self.current_price_business
                self.business_seats -= num_tickets
                self.current_price_business += 200
            else:
                return "Not enough available seats in the business class."
        elif seat_type == 'economy':
            if self.economy_seats >= num_tickets:
                price = self.current_price_economy
                self.economy_seats -= num_tickets
                self.current_price_economy += 100
            else:
                return "Not enough available seats in the economy class."
        else:
            return "Invalid seat type."

        if meal:
            price += 200 * num_tickets

        self.bookings[booking_id] = {
            'seat_type': seat_type,
            'num_tickets': num_tickets,
            'meal': meal,
            'price': price,
            'customer_details': customer_details
        }

        return f"Booking successful. Total price: {price}"
    
    def cancel_booking(self, booking_id, customer_details):
        if booking_id in self.bookings:
            booking = self.bookings[booking_id]
            seat_type = booking['seat_type']
            num_tickets = booking['num_tickets']

            if seat_type == 'business':
                self.business_seats += num_tickets
            elif seat_type == 'economy':
                self.economy_seats += num_tickets
                       
            booking['customer_details'].append(customer_details)  # Add cancellation details

            del self.bookings[booking_id]
            return "Booking cancelled successfully."
        else:
            return "Invalid booking ID"
    
    
    def display_booking_info(self, booking_id):
        if booking_id in self.bookings:
            booking = self.bookings[booking_id]
            print(f"Booking ID: {booking_id}")
            print(f"Seat Type: {booking['seat_type']}")
            print(f"Tickets: {booking['num_tickets']}")
            print(f"Meal: {booking['meal']}")
            print(f"Price: {booking['price']}")
        else:
            print(f"Booking with ID {booking_id} not found.")         
            
    def get_flight_details(self):
        details = f"Flight Number: {self.flight_number}\n"
        details += f"Available seats - Business: {self.business_seats}, Economy: {self.economy_seats}\n"
        details += f"Current price - Business: {self.current_price_business}, Economy: {self.current_price_economy}\n"
        details += "Bookings:\n"
        for booking_id, booking in self.bookings.items():
            details += f"Booking ID: {booking_id}, Seat Type: {booking['seat_type']}, Tickets: {booking['num_tickets']}, Meal: {booking['meal']}, Price: {booking['price']}\n"
            if booking['customer_details']:
                details += "Cancellation Details:\n"
                for cancellation in booking['customer_details']:
                    details += f"Customer Details: {cancellation}\n"
        return details

    def get_ticket_availability(self):
        booked_business_seats = sum(booking['num_tickets'] for booking in self.bookings.values() if booking['seat_type'] == 'business')
        booked_economy_seats = sum(booking['num_tickets'] for booking in self.bookings.values() if booking['seat_type'] == 'economy')
        available_business_seats = self.business_seats - booked_business_seats
        available_economy_seats = self.economy_seats - booked_economy_seats
        availability = f"Flight Number: {self.flight_number}\n"
        availability += f"Available seats - Business: {available_business_seats}, Economy: {available_economy_seats}\n"
        return availability

flight101 = Flight(101)
flight102 = Flight(102)

booking_id1 = 1
booking_id2 = 2
booking_id3 = 3
booking_id4 = 4

flight101.book_ticket(booking_id1, 'economy', 1)
flight101.book_ticket(booking_id2, 'business', 1, meal=True)
flight101.book_ticket(booking_id3, 'business', 2, meal=False)
flight101.book_ticket(booking_id4, 'business', 1, meal=True)

flight101.display_booking_info(1)
print()
flight101.display_booking_info(2)
print()
flight101.display_booking_info(3)
print()
flight101.display_booking_info(4)
print()
print("Summary Of Flight Number 101")
flight101_summary = flight101.get_flight_details()

# Calculate meals booked
meals_booked = sum(booking['num_tickets'] for booking in flight101.bookings.values() if booking['meal'])

# Calculate total seats booked
seats_booked = sum(booking['num_tickets'] for booking in flight101.bookings.values())

# Calculate seat availability
available_business_seats = flight101.business_seats
available_economy_seats = flight101.economy_seats

# Calculate total cost
total_cost = sum(booking['price'] for booking in flight101.bookings.values())

# Add summary information to the flight101_summary string
flight101_summary += f"\nMeals Booked: {meals_booked}"
flight101_summary += f"\nSeats Booked: {seats_booked}"
flight101_summary += f"\nSeat Availability - Business: {available_business_seats}, Economy: {available_economy_seats}"
flight101_summary += f"\nTotal Cost: {total_cost}"

print(flight101_summary)
print('*'*50)
print("Thank you for using AIR INDIA. Have a nice day!")
print("Thank you for using the flight reservation system!")
print('='*50)
print()
flight102.book_ticket(booking_id1, 'economy', 2)
flight102.book_ticket(booking_id2, 'business', 3, meal=True)
flight102.book_ticket(booking_id3, 'business', 2, meal=False)
flight102.book_ticket(booking_id4, 'business', 1, meal=True)

flight102.display_booking_info(1)
print()
flight102.display_booking_info(2)
print()
flight102.display_booking_info(3)
print()
flight102.display_booking_info(4)
print()

print("Summary Of Flight Number 102")
flight102_summary = flight102.get_flight_details()

# Calculate meals booked
meals_booked = sum(booking['num_tickets'] for booking in flight102.bookings.values() if booking['meal'])

# Calculate total seats booked
seats_booked = sum(booking['num_tickets'] for booking in flight102.bookings.values())

# Calculate seat availability
available_business_seats = flight102.business_seats
available_economy_seats = flight102.economy_seats

# Calculate total cost
total_cost = sum(booking['price'] for booking in flight102.bookings.values())

# Add summary information to the flight101_summary string
flight102_summary += f"\nMeals Booked: {meals_booked}"
flight102_summary += f"\nSeats Booked: {seats_booked}"
flight102_summary += f"\nSeat Availability - Business: {available_business_seats}, Economy: {available_economy_seats}"
flight102_summary += f"\nTotal Cost: {total_cost}"
print()

print(flight102_summary)

print('*'*50)
print("Thank you for using AIR INDIA. Have a nice day!")
print("Thank you for using the flight reservation system!")
print('='*50)
```
